# Reconstrução por Interpolação

```python
import numpy as np # operações matemáticas e matriciais
import matplotlib as mpl # funções para apresentação de gráficos e imagens
import matplotlib.pyplot as plt # visualização das imagens e histogramas
from skimage import io # leitura e escrita de imagens
from skimage import transform # funções de transformação
```

## Lendo a imagem a ser reamostrada

```python
img = io.imread('data/images/small_image.png')
print('Dimensões da imagem:', img.shape) # Percebam que quarta dimensão corresponde ao canal "alpha" (transparência)

figsize = (5,5)
plt.figure(figsize = figsize)
plt.imshow(img, interpolation=None), plt.title('Imagem Original') # Indicando que o imshow não deve interpolar a imagem (por default interpolação bilinear)
plt.show()
```

## Reamostrando: dobrando a altura e largura da imagem (x2)

```python
img_nearest = transform.rescale(img, 2, 0, multichannel=True)
img_bilinear = transform.rescale(img, 2, 1, multichannel=True)
img_bicubic = transform.rescale(img, 2, 3, multichannel=True)
print('Dimensões das imagens:', img_nearest.shape)

figsize = (20,10)
plt.figure(figsize = figsize)
plt.subplot(1,3,1), plt.imshow(img_nearest, interpolation=None), plt.title('Vizinho mais próximo')
plt.subplot(1,3,2), plt.imshow(img_bilinear, interpolation=None), plt.title('Bilinear')
plt.subplot(1,3,3), plt.imshow(img_bicubic, interpolation=None), plt.title('Bicubico')
plt.show()
```
