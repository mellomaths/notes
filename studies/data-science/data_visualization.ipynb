We can also reset the index by using the method `reset_index()`. This way we can turn the languages (en, fr, es, ...) into a column in the Dataframe and a counter as indexes (0, 1, 2, ...).

### Using Seaborn catplot

Using catplot you can specify `kind="count"` saving you from all the trouble of doing:

```python
language_count = tmdb_movies
    .original_language
    .value_counts()
    .to_frame()
    .reset_index()
sns.barplot(x="original_language", y="count", data=language_count)
```

````python
sns.catplot(x="original_language", kind="count",data=tmdb_movies)
````

The difference between using the first approach (barplot) and the second (catplot) is that:

- using the barplot approach, as you can see, we already have an order from the most used language to the least because we used the `value_counts()` method and turned its result into a Dataframe.

You can see that, in this case, the Bar Chart makes it hard to see all the languages used in movies, difficulting the visualization.

## Making a more legible chart

We can start by separating how many times the english language appears from how many times any other language appears.

### Separating English language count from others

````python
# Getting the value count for the column original_language again
language_total_count = tmdb_movies.original_language.value_counts()

# Sum of all language counts
language_total_count_all = language_total_count.sum()

# Getting how many times the english language appears
language_total_count_en = language_total_count.loc["en"]

# Getting how many times any language except english appears
language_total_count_all_except_en = language_total_count_all - language_total_count_en


print(f"English appears: {language_total_count_en}")
print(f"Any language except english appears: {language_total_count_all_except_en}")
````

````python
# Create a dictionary with all data collected
language_data = {
    "language": ["english", "others"],
    "total": [language_total_count_en, language_total_count_all_except_en]
}

# Creating a DataFrame using the dictionary above
language_data = pd.DataFrame(language_data)
language_data.head()
````

````python
sns.barplot(x="language", y="total", data=language_data)
````

````python
plt.pie(language_data["total"], labels=language_data["language"])
````

With the Pie Chart, it is not possible to determine and compare magnitude. You can't know how much bigger is the english usage compared to others.

### Giving more details to other language appearances

````python
# Querying all movies in database with original language equals 'en'
tmdb_movies_en = tmdb_movies.query("original_language == 'en'")
tmdb_movies_en.head()
````

````python
# Querying all movies in database with original language different than 'en'
tmdb_movies_not_en = tmdb_movies.query("original_language != 'en'")
tmdb_movies_not_en.head()
````

````python
# Checking if english appears and if the query worked 
#   by counting every language other than english
movies_count_other_languages = tmdb_movies_not_en.original_language.value_counts()
movies_count_other_languages
````

````python
sns.catplot(x="original_language", kind="count", data=tmdb_movies_not_en)
````

It is still difficult to understand this chart. To make it better, we can use the `aspect` (extending the chart width), `order` (giving an order of appearance to the X axis), and `palette` (changing bar colors to be sequential) parameters.

You can access [this tutorial](https://seaborn.pydata.org/tutorial/color_palettes.html) on choosing color palettes in the Seaborn documentation.

````python
sns.catplot(
    x="original_language", 
    kind="count", 
    data=tmdb_movies_not_en, 
    aspect=2, 
    order=movies_count_other_languages.index,
    palette="GnBu_d")
# Aspect parameter is used to handle chart width
# Aspect ratio of each facet, so that aspect * height gives the width of each facet in inches

# Order parameter receives the serie responsible to be the actual order that appears in the X axis
# We can use movies_count_other_languages series that is already ordered, and access the .index method
# Order to plot the categorical levels in, otherwise the levels are inferred from the data objects.

# Palette paramete - Colors to use for the different levels of the hue variable. 
# Should be something that can be interpreted by color_palette(), or a dictionary mapping hue levels to matplotlib colors.
````
