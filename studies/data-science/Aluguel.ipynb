## Selecionado Imóveis Residenciais 

```python
imoveis_residenciais = list(aluguel['Tipo'].drop_duplicates())
imoveis_residenciais
```

```python
imoveis_residenciais = [
    'Quitinete',
    'Casa',
    'Apartamento',
    'Casa de Condomínio',
    'Casa de Vila'
]
imoveis_residenciais
```

```python
aluguel.head(10)
```

```python
# Criando uma pd.Series com valores booleanos 
# -- para selecionar as linhas que se referem a um Imóvel Residencial
selecao_imoveis_residenciais = aluguel['Tipo'].isin(imoveis_residenciais)
selecao_imoveis_residenciais.head(10)
```

```python
# Selecionando apenas imóveis residenciais
aluguel_residencial = aluguel[selecao_imoveis_residenciais]
aluguel_residencial
```

```python
list(aluguel_residencial['Tipo'].drop_duplicates())
```

```python
print(f"Quantidade Alugueis de qualquer tipo: {aluguel.shape[0]}")
print(f"Quantidade Alugueis do tipo Residencial: {aluguel_residencial.shape[0]}")
```

```python
# Redefinindo os índices do pd.DataFrame de Imóveis Residenciais
aluguel_residencial.index = range(aluguel_residencial.shape[0])
aluguel_residencial.head(10)
```

### Exportando a Base de Dados

```python
aluguel_residencial.to_csv("data/aluguel_residencial.csv", sep=";")
```
